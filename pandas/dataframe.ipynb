## Reindexing DataFrame

```python
dframe = DataFrame(np.random.rand(25).reshape((5, 5)), index=['A', 'B', 'F', 'D', 'E'], 
                   columns=['col1', 'col2', 'col3', 'col4', 'col5']) 
# To reindex rows, call reindex and pass in a new list
dframe.reindex([x if x != 'F' else 'C' for x in dframe.index])

# To reindex columns, call reindex and pass in new list 
# with param columns.
# Again, remember, missing values from previous list
# will not be included in the new list, and new values
# will have NaNs
dframe.reindex(columns=['col1', 'col2', 'col3', 'col4', 'lmnop'])
```

## Difference in loc, iloc, ix

```python
# loc: loc works on labels in the index.
# iloc: iloc works on the positions in the index (so it only takes integers).
# ix: ix usually tries to behave like loc but falls back to behaving like iloc if the label is not in the index.
# It's important to note some subtleties that can make ix slightly tricky to use:

# 1. if the index is of integer type, ix will only use label-based indexing and not fall back to position-based 
# indexing. If the label is not in the index, an error is raised. 


# 2. if the index does not contain only integers, then given an integer, ix will immediately use position-based indexing
# rather than label-based indexing. If however ix is given another type (e.g. a string), it can use label-based 
# indexing.
```

## Data Alignment


```python
df = DataFrame(np.arange(4).reshape((2, 2)), columns=list('AB'), index=list('12'))
df

df2 = DataFrame(np.arange(9).reshape((3, 3)), columns=list('ABC'), index=list('156'))
df2

# only values where the (index, column) are both same in both
# dataframes are added, rest all of them are NaNs
df + df2

# If we don't want NaNs, we can pass in a default value
# remember there might still be NaNs, where the row
# and col are mutually exclusive
df.add(df2, fill_value=0)
```

## Summary Statistics

```python
arr = np.array([[1, 2, np.nan], [3, 4, np.nan]])
df = DataFrame(arr, index=[1, 2], columns=['One', 'Two', 'Three'])

# Sum elements column wise
# NaNs are considered 0
df.sum()

# returns minimum of each column
# here NaNs are not treated as zero
# min of column with all NaNs is NaN
df.min()  # or df.max() for maximum

# Gives the index of the minimum columns
# returns NaN for column with all NaNs
df.idxmin()  # or df.idxmax() for maximum

# Cumulative sum, adds number after every
# row
# eg. 
# A[0, 0]                 A[0, 1]           A[0, 2]
# A[0, 0]+A[1, 0]         A[0, 1] + A[1, 1] A[0, 2] + A[1, 2] 
# A[0, 0]+A[1, 0]+A[2, 0]   etc..
df.cumsum()

# Describe the DataFrame like summary statistics
df.describe()


# Covariance: a measure of linear relationship between two variables.
# positive indicates direct or increasing covariance, negative indicates
# decreasing linear relationship.

# When covariance is near zero it indicates there's no relationship
# between the variables.

# Covariance provides DIRECTIONS: positive covariance, or negative,
# but Correlation provides DIRECTIONS and STRENGTH, how strongly
# correlated variables are

# Covariance has no limits, Correlation has, that's +1 and -1.

# Correlation strength does not imply statistical significance. 

# Correlation works only with LINEAR DATA. For non-linear, it doesnt
# work.
```

## Filling missing data

```python
na = np.nan
```

```python
df = DataFrame([[1, 2, 2], [3, 4, na], [na, na, na]])
df

# True if entry is NaN
df.isnull()

# drop entire row even if a single entry is NaN
df.dropna()

# drop entire row only if all the entries are NaNs
df.dropna(how='all')

# instead of rows, target columns
df.dropna(axis=1)

# only want rows that have more than '2' 
# data points
df.dropna(thresh=2)

# fill all the NaNs with some value
df.fillna(1)

# can specify different for different columns
# pass in a dict where key is column and value is
# the value you want to insert
df.fillna({0: np.mean(df[0]), 1: np.mean(df[1]), 2: np.mean(df[2])})
```

## Indexing heirarchy

```python
df = DataFrame(np.arange(16).reshape((4, 4)),
               index=[[1, 1, 2, 2], ['a', 'b', 'a', 'c']],
               columns=[['NY', 'NY', 'SF', 'SF'],
                        ['alpha', 'beta', 'gamma', 'delta']])
df

# Can name your index so you  
# won't get lost
df.index.names = ['one_two', 'a_b']
df.index

# Same applies to columns
df.columns.names = ['CITIES', 'GREEK LETTERS']
df.columns
df

# Swap levels (axis=1 for columns)
df.swaplevel('CITIES', 'GREEK LETTERS', axis=1)

# Sort levels
# df.sortlevel(0) will sort level 0 index at axis=0
# df.sortlevel(0, axis=1) will do the same with column
df.sortlevel(1, axis=1)

# Aggregate similar columns and sum them 
df.sum(level='CITIES', axis=1)
```
