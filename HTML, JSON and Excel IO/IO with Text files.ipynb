```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from os import getcwd as cwd
file_loc = cwd() + '\\HTML, JSON and Excel IO\\text_data.csv'
```

```python
# read from CSV
# automatically set's first row as column names
df = pd.read_csv(file_loc)
df
```

```python
# if you don't want first row to be header
df = pd.read_csv(file_loc, header=None)
df
# we'll have 0, 1, 2, 3 as columns
```

```python
# Generic reader
df = pd.read_table(file_loc, sep=',')
df  # Will get same table
```

```python
# Just read few rows, specified by nrows
pd.read_csv(file_loc, nrows=3)
```

```python
# Save df to a csv file
# can save with different separator
# specify which column in columns= field
df.to_csv('my_df_out.csv', sep=',', columns=['name'])  
df2 = pd.read_csv('my_df_out.csv')
df2
```
